$$ A = \left[ \begin{array}{ccc}
        10 & 2 & 2 \\
        -3 & 1 & 4 \\
        1 & 0 & 3 \\
         5 & -1 & 5 
    \end{array} \right] $$

a)	Multiplique la fila 1 por 1/4.

b)	Sume 3 veces la fila 1 a la fila 2.

c)	Sume (-1) veces la fila 1 a la fila 3.

d)	Sume (-5) veces la columna 1 a la columna 3.

e)	Intercambie las filas 2 y 4.

f)	Sea $B$ la matriz resultante de aplicar $a) - e)$ secuencialmente sobre la matriz $A$. Halle la traza de $B \times B^T$

g)	Halle la matriz inversa de $B \times B^T$

```python
# Definición de la matriz
A = np.array([[10, 2, 2],[-3, 1, 4], [1, 0, 3], [5, -1, 5]], dtype=float)
print(A)
```

### a) Multiplique la fila 1 por 1/4.

```python
# a)
A[0] = A[0]*(1/4)
print(f"Multiplique la fila 1 por 1/4: \n {A}")
```

### b) Sume 3 veces la fila 1 a la fila 2.

```python
# b)
A[1] = A[1] + 3*A[0]
print(f"Sume 3 veces la fila 1 a la fila 2: \n {A}")
```

### c) Sume (-1) veces la fila 1 a la fila 3.

```python
# c)
A[2] = A[2]-A[0]
print(f"Sume (-1) veces la fila 1 a la fila 3: \n {A}")
```

### d) Sume (-5) veces la columna 1 a la columna 3.

```python
# d)
A[:,2] = A[:,2] - 5*A[:,0]
print(f"Sume (-5) veces la columna 1 a la columna 3: \n {A}")
```

### e) Intercambie las filas 2 y 4.

```python
# e)
fila2 = np.array(A[1])
A[1] = A[3]
A[3] = fila2
B = A
print(f"Intercambie las filas 2 y 4: \n {A}")
print(f"La matriz resultante de todas estas operaciones es B: \n {B}")
```

### f) Sea  *B*  la matriz resultante de aplicar  *a)-e)*  secuencialmente sobre la matriz  *A*. Halle la traza de *B×B<sup>T</sup>*.

```python
# f)
transpB = np.transpose(B)
print(f"B^T (B transpuesta) es igual a \n {transpB}")
multip = np.matmul(B, transpB)
print(f"B x B^T es igual a \n {multip}")
respf = multip.trace()
print(f"La traza de B x B^T es {respf}")
```

### g) Halle la matriz inversa de *B×B<sup>T</sup>*.

En este caso, nos damos cuenta que el determinante de la matriz *B×B<sup>T</sup>* es un número que es prácticamete cero (orden de 10<sup>-11</sup>). Esto quiere decir que la inversa de esta matriz no está definida. Sin embargo, se puede notar que *numpy* encuentra una matriz inversa teniendo en cuenta el valor indicado por el determinante, resultado que está completamente fuera de proporciones. 

```python
# g)
respg = np.linalg.inv(multip)
red = np.round(respg, 3)
det = np.linalg.det(multip)
print(f"Determinante: {det}")
print(f"La inversa de \n {multip} \n es \n {red}")
```

Se prueba que la matriz encontrada por *numpy* no corresponde a un resultado correcto sabiendo que el producto matricial entre una matriz *A* y su inversa *A<sup>-1</sup>* debe ser igual a la matriz identidad (matriz cuadrada de la misma dimensión que *A* cuya diagonal contiene 1s).

```python
f = np.matmul(multip, respg)
print(f)
```

Como se puede notar, el producto matricial entre *multip* y *respg*, que corresponden respectivamente a las matrices B×B<sup>T</sup> y (B×B<sup>T</sup>)<sup>-1</sup> **no** es igual a la matriz identidad, así probando el error en el que se cae al invertir una matriz cuyo determinante es prácticamente cero usando la librería *numpy*.

### a) 
Se define una función cuyo parámetro es un número entero, que retorna 1 en caso de que el número ingresado sea un número capicúa y 0 en caso contrario.

```python
def capicua(num):
    
    # Primero se toma el número y se convierte en un string para poder acceder a cada uno de sus dígitos individualmente.
    string = str(num)
    
    # Se crea un string auxiliar para guardar el valor del número ingresado por parámetro, pero escrito al revés.
    string1 = ""
    
    # Se recorre el número que ingresó por parámetro, dígito por dígito.
    for i in range(len(string)):
        # Se toma cada dígito del número (de derecha a izquierda) y se va agregando al string auxiliar creado anteriormente.
        string1 = string1 + string[len(string)-1-i]
    # Al terminar el recorrido, string1 es el mismo string, escrito al revés
    
    # Se comparan ambas cadenas de carácteres y se retorna 1 o 0 dependiendo si son exactamente lo mismo, lo que indicaría
    # que efectivamente el número ingresado es un número capicúa.
    if string == string1:
        return 1
    else:
        return 0
```

### b)
Se define una función que recibe como parámetro un número entero y devuelve la cantidad de
números capicúa entre 0 y dicho número (incluyendo dicho número).

```python
def cuantosCapicua(num):
    
    # Se define una variable para contar cuántos números capicúa se van encontrando en un recorrido.
    contador = 0
    
    # Se recorren los números entre 0 (por defecto) y el número ingresado por parámetro. Se le suma 1 a dicho número para 
    # así tener este en cuenta.
    for i in range(num+1):
        
        # Si la función retorna 1, se le suma 1 al contador. 
        if capicua(i) == 1:
            contador += 1
            
    return contador

# Se usa la función time para contar cuánto tiempo (en segundos) se demora la función recientemente creada en correr. 
# Adicionalmente, se crea una variable para ingresarla a la función como parámetro. En este caso, se prueba con 10,000,000.
prueba = 10**7
start = time.time()
resp = cuantosCapicua(prueba)
end = time.time()

tiempo = round(end-start, 3)
    
print(f"Hay {resp} número(s) capicúa entre 0 y {prueba}\nTiempo de ejecución: {tiempo}s")
```

## Punto 3
Se define una función $f(a, b, rep)$ que lee dos arreglos de números enteros y muestra en consola una lista
ordenada, con los elementos de ambos arreglos. Se usó la función *sorted*, cuya documentación se encuentra [aquí](https://www.w3schools.com/python/ref_func_sorted.asp#:~:text=The%20sorted()%20function%20returns,string%20values%20AND%20numeric%20values.).

```python
# Parámetros de la función:
# a: Uno de los arreglos.
# b: Otro arreglo.
# rep: Variable de control que determina si el arreglo resultante tiene o no elementos repetidos.

def f(a, b, rep):
    
    # Si se quiere que el arreglo resultante tenga elementos repetidos
    if rep == True:
        for item in b:
            # Se agrega el item del arreglo 'b' al arreglo 'a'.
            a.append(item)
        # Al finalizar el recorrido, se ordena la lista en orden ascendente. 
        ordenada = sorted(a)
    else: # Si se quiere que el arreglo resultante no tenga elementos repetidos
        for item in b:
            # Se agrega el item del arreglo 'b' al arreglo 'a'.
            a.append(item)
        # Se convierte la lista 'a' en un set. Al hacer esto, todos los duplicados posibles se eliminan. 
        a = set(a)
        # Se ordena el set en orden ascendente. 
        ordenada = sorted(a)
        
    return ordenada
    
    
lista11 = [3, 9, 5, 1, 7]
lista12 = [8, 4, 6, 2, 1, 5, 7]

lista21 = [3, 20, 17, 4, 22, 50]
lista22 = [18, 21, 17, 34, 22, 101, 85]

lista31 = [5, 6, 23, 76, 243, 68, 193, 45]
lista32 = [1, 0, 34, 83, 75, -5, -10, -10]
```

Se prueba la función *f* con los arreglos **lista21** y **lista22**.

```python
f(lista21,lista22,True)
```

```python
f(lista21,lista22,False)
```

Se prueba la función *f* con los arreglos **lista31** y **lista32**.

```python
f(lista31, lista32, True)
```

```python
f(lista31, lista32, False)
```

Se puede decir que la función funciona porque al entregarle listas con ítems repetidos, muestra correctamente estos ítems repetidos o no dependiendo si el parámetro $rep$ es **True** o **False**. Además, la función siempre entrega la lista resultante de unir los dos parámetros $a$ o $b$ ordenada en orden ascendente.

## Punto 5



De acuerdo al enunciado del problema, se sabe que los productos químicos deben pasar por dos máquinas antes de terminar su producción. Se sabe que estos productos químicos se demoran cierta cantidad de horas por tonelada en cada máquina. Por lo tanto, se tienen dos funciones objetivos a optimizar (en este caso, lograr que las máquinas se usen a su máxima capacidad): una para cada máquina . 

Teniendo en cuenta los datos entregados en el enunciado, y asignando las variables de decisión $x_1$, $x_2$ y $x_3$ como las toneladas de *A*, *B* y *C* respectivamente, las funciones objetivo son:

Máquina X: $f_{X}(x_1,x_2,x_3) = 2x_{1} + 3x_{2} + 4x_{3}$

Máquina Y: $f_{Y}(x_1,x_2,x_3) = 2x_{1} + 2x_{2} + 3x_{3}$

Sabiendo que las horas disponibles de la máquina X son 80, mientras que las horas disponibles de la máquina Y son 60, y además que las toneladas de cada producto químico siempre son positivas, se ponen las siguientes restricciones:

(1) $2x_{1} + 3x_{2} + 4x_{3} = 80$

(2) $2x_{1} + 2x_{2} + 3x_{3} = 60$

(3) $x_{1}, x_{2}, x_{3} \geq 0$

### a) Se usa la librería *sympy* para plantear el sistema de ecuaciones del problema. Información sobre cómo usar la librería *sympy* [aquí](https://apmonitor.com/che263/index.php/Main/PythonSolveEquations).

```python
sym.init_printing()
x1,x2,x3 = sym.symbols('x1,x2,x3')

fx = sym.Eq(2*x1+3*x2+4*x3,80)
fy = sym.Eq(2*x1+2*x2+3*x3,60)
```

### b) Se usa la librería *sympy* para resolver el sistema de ecuaciones correspondiente a este problema. 

```python
# Se resuelven las ecuaciones fx y fy con respecto a las variables x1, x2 y x3. Como solo se tienen dos ecuaciones
# y dos incógnitas, la solución queda en términos de la variable faltante. Es decir, se puede graficar como una curva
# paramétrica.

sym.solve([fx,fy],(x1,x2,x3))
```

De acuerdo a esta solución, se deben producir $10 - \frac{C}{2}$ (a) toneladas de *A* y $20-C$ (b) toneladas de *B*, donde las toneladas de *C* son un parámetro libre. Esto quiere decir que no importa cuántas toneladas se usen de *C*, las máquinas siempre van a operar a toda su capacidad, y por supuesto la cantidad de toneladas de *A* y *B* están sujetas a la cantidad de toneladas de *C*, siguiendo las relaciones expresadas en las ecuaciones a y b.

### c) Se dibuja en un espacio cartesiano de 3 dimensiones las soluciones al problema, que en este caso, están dadas por una recta en el espacio tridimensional.

Se encontró información sobre cómo graficar funciones paramétricas en espacio tridimensional en [esta discusión](https://stackoverflow.com/questions/37711538/matplotlib-3d-axes-ticks-labels-and-latex) de StackOverflow.

```python
# Se crea la figura en espacio 3d.
fig = mat.figure()
ax = fig.gca(projection='3d')

# Se crea un vector equitativamente espaciado con valores entre 0 y 30 para mostrar estos valores que puede tomar
# la variable x3, es decir, las toneladas de C.
x3 = np.linspace(0, 30)

# Se definen las toneladas de A y B respectivamente de acuerdo a la solución al sistema de ecuaciones
# planteado anteriormente. 
x1 = 10-x3/2
x2 = 20-x3

# Se dibuja la gráfica y la leyenda de la misma. 
ax.plot(x1, x2, x3, label="Región Factible")
ax.legend()

# Se definen los valores entre los que nos interesa mirar la variación de las toneladas de A y B. 
ax.set_xlim([0, 15])
ax.set_ylim([0, 15])

# Se le pone una etiqueta a cada eje coordenado. 
ax.set_xlabel('Tons de A')
ax.set_ylabel('Tons de B')
ax.set_zlabel('Tons de C')
```
