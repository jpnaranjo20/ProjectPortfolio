## Punto 2

```python
# Orden del filtro
N = 50

# La entrada es la señal antes de ser filtrada.
x = yr

# La salida es la señal después de ser filtrada.
y = yf

# Se define el vector Y para que se tomen los datos a partir de la posición N.
Y = y[N:]

# Se construye la matriz H
H = np.zeros((len(Y),N+1))
for i in range(len(Y)):
    for j in range(N+1):
        H[i,j] = x[N+(i-j)]

# Se estiman los parámetros
theta = LA.inv((H.T)@H)@(H.T)@(Y)

print(f'theta: \n {theta}')
```

Para calcular los coeficientes del filtro FIR de orden $N$, hay que tener en cuenta que la ecuación que define la respuesta de dicho filtro es:
$$y(k) = \sum_{i=0}^N \theta_i x(k-i)$$
en donde los $\theta_i$ representan a los coeficientes que filtran la señal de entrada $x(k)$. 

Como se sabe del probelma de regresión lineal usual, para estimar los parámetros $\theta$, conociendo la salida del filtro y su entrada, la ecuación que define los parámetros de la regresión es:
$$\theta^* = (H^TH)^{-1}H^TY$$
en donde el vector $Y$ está definido como

$$Y = \begin{bmatrix}
y[N]\\
y[N+1]\\
y[N+2]\\
\vdots
\end{bmatrix}$$

mientras que la matriz $H$ está dada por la expresión:

$$H = \begin{bmatrix} 
    x[N] & x[N-1] & \dots & x[0]\\
    x[N+1] & x[N] & \dots & x[1]\\
    x[N+2] & x[N+1] & \dots & x[2]\\
    \vdots & \vdots & \ddots & \vdots\\
    \end{bmatrix}$$

De ahí se sabe que la matriz $H$ va a ser de tamaño $(tamY-N)$x$(N+1)$ y que el vector $Y$ debe tener tamaño $(tamY-N)$x$1$


## Punto 3

Se encuentra el espectro de los coeficientes recientemente encontrados para determinar las frecuencias que se anulan y las que el filtro deja pasar.

```python
fouriertheta = fft(theta)
fs = fftfreq(np.size(theta), 1/samprate2)

plt.figure()

plt.plot(fs, np.abs(fouriertheta), color='r')
plt.grid()
plt.xlabel('$f$ (Hz)')
plt.ylabel(r'|$\theta(f)$|')
plt.title(r"Transformada de Fourier de los coeficientes $\theta$")
```

Se puede notar que el filtro tiene el efecto de anular todas ls frecuencias superiores a los 2000 Hz (aproximadamente). Se puede argumentar que este es el resultado esperado, porque el ruido que se ve en la señal 1 está ubicado en la frecuencia de los 3000 Hz. Después de pasar por el filtro, el ruido se elimina porque todas las frecuencias superiores a los 2000 Hz son eliminadas por el filtro FIR.

Comparando con la respuesta de frecuencia de la señal filtrada mostrada en el punto 1, se puede afirmar que los coeficientes encontrados son correctos pues su respuesta de frecuencia solo deja pasar las frecuencias menores a los 2000 Hz (aproximadamente) y en la respuesta de frecuencia de la señal filtrada del punto 1 solo hay frecuencias diferentes de cero en esta banda (-2000 Hz y 2000 Hz).
