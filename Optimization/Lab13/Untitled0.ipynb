```python
import numpy as np
import random
from numpy import linalg as LA
```

```python
def funcion(m,n,U,X,V):
    valor = 0
  
    for i in range(n):
        for j in range(m):
            valor = valor + U[i,j]**2 * LA.norm(X[i,:]-V[j,:])**2
        
    return valor
```

```python
k = 0
n=200
m=3
epsilon = 0.001
mu = 10
sigma = 10


#Valores iniciales
U1 = np.ones((n,1)) # Columna de 1s
U = np.zeros((n,m-1))
U = np.column_stack((U1,U))
X = np.random.normal(mu, sigma, size=(n, 2))
V = np.array([[1,2],[18,20],[-2,10]])

print(U1)



valor = funcion(m,n,U,X,V)
print(valor)

# while valor >= epsilon:

  # valor_funcion = funcion(actual) - funcion(actual-1)
```
