```python
import numpy as np
import sympy
from sympy import *
import itertools
from itertools import combinations
import scipy as sp
from scipy.optimize import linprog
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def sol_primal(A,b,c,f):

    vectorsol = f[0]
    valoropti = f[1]
    
    print(f'vectorsol: \n {vectorsol}')
    print(f'valoropti: \n {valoropti}')
    
    c = -1*c
    
    A_dual = A.transpose()
    c_dual = b.transpose()
    b_dual = c.transpose()
    
    print(f'A_dual: \n {A_dual}')
    print(f'c_dual: \n {c_dual}')
    print(f'b_dual: \n {b_dual}')
    
    eqs = []#list()
    valnul = []#list()
    
    # Primera condición de hc: Si no son activas, se toma el índice de la variable que se iguala a cero.
    for i in range(len(vectorsol)):
        if (vectorsol[i] != 0):
            eqs.append(i)
            print(f'eqs:{eqs}')
    
    # Primera condición de hc: Se reemplaza el valor solución en las restricciones del dual. El resultado queda
    # guardado en Ma_dec.
    Ma_dec = A_dual.dot(vectorsol).transpose()
    print(f'Ma_dec: \n {Ma_dec}')
    
    sistema = A.copy()[eqs,:]
    igualdad = b.copy()[eqs,:] 
    
    print(f'sistema: \n {sistema}')
    print(f'igualdad: \n {igualdad}')
    
#     comp = [] #list()
    
    # Se determina si las restricciones de la 1ra cond de hc son activas o inactivas. False -> Activa, True -> Inactiva
    comp = [Ma_dec[z]>b_dual[z] for z in range(len(Ma_dec))]
    
    print(f'comp: \n {comp}')
    
    for j in range(len(comp)):
        # Si la resitricción j es inactiva, significa que la variable xj = 0. Se agrega el índice a una lista.  
        if (comp[j]):
             valnul.append(j)
    
    print(f'valnul: \n {valnul}')
    
    # Se eliminan las columnas del sistema y de la función objetivo donde la variable es nula.
    # Se concatena el sistema resultante con el vector b (matriz aumentada). 
    sistema = np.delete(sistema, valnul,axis=1)
    c = np.delete(c, valnul, axis=1)
    Ab = np.c_[sistema,igualdad]
    
    print(f'sistema: \n {sistema}')
    print(f'c: \n {c}')
    print(f'Ab: \n {Ab}')
    
    # Se toman los indices de las columnas pivotes, para saber qué variables no son nulas. 
    _ , inds = sympy.Matrix(Ab).rref()
    inds = np.array(inds)
    print(f'inds: \n {inds}')
    A_temp = sistema[inds]
    b_temp = igualdad[inds]
    print(f'A_temp: \n {A_temp}')
    print(f'b_temp: \n {b_temp}')
    
    # Si al eliminar las variables nulas, queda una matriz cuadrada, se encuentra la solución.  
    if (len(inds) == len(b_temp)):
        print('len(inds) == len(b_temp)')
        x = np.linalg.inv(A_temp) @ b_temp
        print(f'x: \n {x}')
    else: 
    # De lo contrario, se concatenan 
        print('len(inds) != len(b_temp)')
        A_temp = np.concatenate((A_temp, [c]), axis=0)
        print(f'A_temp: \n {A_temp}')
        b_temp = np.concatenate((b_temp,[valoropti]), axis=0)
        print(f'b_temp: \n {b_temp}')
        x = np.linalg.inv(A_temp)@b_temp
        print(f'x: \n {x}')
        
    fx=-1*c.dot(x)
    x=np.insert(x, valnul, 0)
    print(f'x_insert: \n {x}')
    
    return fx, x
```

```python
# Utilidad del primal Alternativa 2a
#Coeficientes que acompañan a las restricciones
An=np.array([[7/10, 1],[1/2, 5/6],[1, 2/3],[1/10,1/4]])

#Cotas superiores de las restricciones
b=np.array([[630],[600],[708],[135]])

#Coeficientes que acompañan a las variables de decisión en la función objetivo
c = np.array([[-14,-9]])

f = np.array([[0, 0, 14, 0],[9912]])

[fx, x] = sol_primal(An, b, c, f)
print(f'La solución al problema original de optimización es x = {x}')
print(f'El valor de la función objetivo evaluada en el valor óptimo es {-1*fx}')
```

```python
# Utilidad del primal Alternativa 2a
#Coeficientes que acompañan a las restricciones
An=np.array([[3,1],[5, 2]])

#Cotas superiores de las restricciones
b=np.array([[4],[7]])

#Coeficientes que acompañan a las variables de decisión en la función objetivo
c = np.array([[-6, -8]])

f = np.array([[0, 6/5],[8.4]])

[fx, x] = sol_primal(An, b, c, f)
print(f'La solución al problema original de optimización es x = {x}')
print(f'El valor de la función objetivo evaluada en el valor óptimo es {-1*fx}')
```
