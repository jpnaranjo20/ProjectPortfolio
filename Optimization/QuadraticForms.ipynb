# Quadratic Forms

```python
import numpy as np
from numpy import inf
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

## Part 1

Consider the quadratic form $f(x) = x^{T}A_{i}x$ and the following matrices:

$ A_1 = \begin{bmatrix}
1 & 0 \\ 
0 & 3 \\ 
\end{bmatrix} $

$A_2 = \begin{bmatrix}
-9 & 0 \\ 
0 & -9 \\ 
\end{bmatrix}$

$A_3 = \begin{bmatrix}
1 & 5 \\ 
5 & 1 \\ 
\end{bmatrix}$

$A_4 = \begin{bmatrix}
5 & 1 \\ 
1 & 5 \\ 
\end{bmatrix} $


### a) Plots

```python
A1 = np.array([[1, 0], [0, 3]])
A2 = np.array([[-9, 0], [0, -9]])
A3 = np.array([[1, 5], [5, 1]])
A4 = np.array([[5, 1], [1, 5]])

matrices = [A1, A2, A3, A4]

# Counter to know what title to give to each plot.
c = 0

# Iteratively draw each plot.
for mat in matrices:
    step = 0.025
    x = np.arange(-1.0, 1.0, step)
    y = np.arange(-1.0, 1.0, step)
    
    X, Y = np.meshgrid(x, y)
    # Z is the quadratic form associated with each matrix
    Z = mat[0, 0] * X**2 + (mat[0, 1] + mat[1, 0]) * X * Y + mat[1, 1] * Y**2

    fig, ax = plt.subplots()
    CS = ax.contour(X, Y, Z)
    ax.clabel(CS, inline=True, fontsize=10)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    c += 1
    
    # These conditionals are declared to give a title to each plot.
    if c == 1:
        ax.set_title(f'Contour of A1')
    elif c == 2:
        ax.set_title(f'Contour of A2')
    elif c == 3:
        ax.set_title(f'Contour of A3')
    elif c == 4:
        ax.set_title(f'Contour of A4')
```

### b) The function $f$ is plotted over the range $[−1, 1]^2$ for each matrix.

```python
# Counter
c = 0

# Iteratively carry out the plotting process.
for mat in matrices:
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    x = y = np.arange(-1, 1, 0.05)
    X, Y = np.meshgrid(x, y)
    # Z is the quadratic form
    Z = mat[0, 0] * X**2 + (mat[0, 1] + mat[1, 0]) * X * Y + mat[1, 1] * Y**2

    ax.plot_surface(X, Y, Z)

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    c += 1
    # The same conditionals are used to know what title to assign to each plot.
    if c == 1:
         plt.title('f(x) with A1')
    elif c == 2:
         plt.title('f(x) with A2')
    elif c == 3:
         plt.title('f(x) with A3')
    elif c == 4:
         plt.title('f(x) with A4')

    plt.show()
```

### c)
The eigenvalues and eigenvectors are found for the matrices $A_1, A_2, A_3, A_4$

It is worth mentioning that, according to the operation of the *eig* function, the eigenvalues are printed as a 1x2 vector (all matrices are 2x2), and each component of this vector represents one of the eigenvalues.

Similarly, according to the operation of the *eig* function, the eigenvectors are printed as a matrix, where each column is one of the eigenvectors. The eigenvector of the i-th column is associated with the i-th component of the eigenvalue vector returned by the *eig* function.

```python
c = 0

# Eigenvalues and eigenvectors are found iteratively.
for mat in matrices:
    c += 1
    l, v = LA.eig(mat)
    if c == 1:
        print(f'Matrix A1: \n {mat}')
        print(f'Eigenvalues: \n {l}')
        print(f'Eigenvectors: \n {v}')
    elif c == 2:
        print(f'\nMatrix A2: \n {mat}')
        print(f'Eigenvalues: \n {l}')
        print(f'Eigenvectors: \n {v}')
    elif c == 3:
        print(f'\nMatrix A3: \n {mat}')
        print(f'Eigenvalues: \n {l}')
        print(f'Eigenvectors: \n {v}')
    elif c == 4:
        print(f'\nMatrix A4: \n {mat}')
        print(f'Eigenvalues: \n {l}')
        print(f'Eigenvectors: \n {v}')
```

The eigenvectors of each matrix are related to their respective contour plot in the sense that they serve as a "new coordinate axis". That is, the eigenvectors function as a new plane $\lambda_1\lambda_2$, which can be called the $\lambda_1\lambda_2$ plane, and which lies on the $xy$ plane but rotated according to the values of the eigenvectors. In other words, the new coordinate axis is constructed by the two eigenvectors of each matrix and is rotated relative to the original coordinate axis, according to the components of each eigenvector.

The eigenvalues of each matrix are related to their respective contour plot depending on whether the quadratic form is an ellipse or a hyperbola.

For ellipses, $\pm\sqrt{\lambda_1}$ represent the vertices on the $\lambda_1$ axis of the ellipse, while $\pm\sqrt{\lambda_2}$ indicate the vertices on the $\lambda_2$ axis of the ellipse. These distances are measured with respect to the $\lambda_1\lambda_2$ plane.

For the unique hyperbola, corresponding to matrix $A_3$, the eigenvalues of this matrix indicate the position (also measured relative to the $\lambda_1\lambda_2$ plane) of the vertices of the hyperbola.

## Part 2

### We consider the quadratic form $f(x) = x^{T}A_{\rho}x$ such that:
$$ A_\rho = \begin{bmatrix}
1 & \rho \\ 
\rho & 1 \\ 
\end{bmatrix} $$


### a) 
Graph of the function using matplotlib's *plot_surface*.

```python
# A list with all possible values of rho is declared to iteratively plot.
rho = [0, -0.25, 0.25, -0.75, 0.75, 1, -1, 2, -2]

for p in rho:
    A = np.array([[1, p],[p, 1]])
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    x = y = np.arange(-1, 1, 0.05)
    X, Y = np.meshgrid(x, y)
    # The quadratic form varies depending on the value of rho.
    Z = A[0,0]*X**2 + (A[0,1] + A[1,0])*X*Y + A[1,1]*Y**2

    ax.plot_surface(X, Y, Z)

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    plt.title(f'f(x) with p = {p}')
    plt.show()
```

### b)  
Contour plot using the *contourf* function and *colorbar* function from matplotlib.

```python
# Lower and upper limits for the axes on which each contour curve will be plotted.
inf = -3
sup = 3

for p in rho:
    
    # The matrix changes according to the value of rho.
    A = np.array([[1, p],[p, 1]])

    delta = 0.025

    x = y = np.arange(inf, sup, delta)
    X, Y = np.meshgrid(x, y)
    
    # Quadratic form associated with the matrix.
    Z = A[0,0]*X**2 + (A[0,1] + A[1,0])*X*Y + A[1,1]*Y**2

    fig1, ax2 = plt.subplots(constrained_layout=True)
    CS = ax2.contourf(X, Y, Z, 10, cmap=plt.cm.bone, origin=None)
    ax2.set_title(f'Contour of f(x) with p = {p}')
    ax2.set_xlabel('X')
    ax2.set_ylabel('Y')
    
    # Color bar
    cbar = fig1.colorbar(CS)
    cbar.ax.set_ylabel('Z')
    plt.show()
```

### c) 
Eigenvalues and eigenvectors of the matrix $A_\rho$ for each possible value of $\rho$.

```python
for p in rho:
    A = np.array([[1,p],[p,1]])
    l, v = LA.eig(A)
    print(f'\np = {p}')
    print(f'Matrix A_p: \n {A}')
    print(f'Eigenvalues: \n {l}')
    print(f'Eigenvectors: \n {v}')
```

## Part 3

### We consider the matrices $A_1, A_2, A_3, A_4$ from part 1.

### a) 

It is found that the eigenvectors of each matrix already have unit magnitude. The vectors $\lambda_1v_1$ and $\lambda_2v_2$ are constructed and plotted in the plane.


```python
c=0

# Each matrix finds its eigenvalues and eigenvectors, and these values are saved in different variables.
# Since the eigenvalues are expressed as components of a row vector, reference to the desired value must be made
# through numpy vector notation, and the same applies to obtain the eigenvectors, which are the columns of the parameter v.
for mat in matrices:
    c+=1
    l, v = LA.eig(mat)
    if c==1:
        a1 = l[0]*v[:,0]
        a11 = l[1]*v[:,1]
    elif c==2:
        a2 = l[0]*v[:,0]
        a22 = l[1]*v[:,1]
    elif c==3:
        a3 = l[0]*v[:,0]
        a33 = l[1]*v[:,1]
    elif c==4:
        a4 = l[0]*v[:,0]
        a44 = l[1]*v[:,1]
        
print(f'Vectors of matrix A1: \n v1: {a1} \n v2: {a11}\n')
print(f'Vectors of matrix A2: \n v1: {a2} \n v2: {a22}\n')
print(f'Vectors of matrix A3: \n v1: {a3} \n v2: {a33}\n')
print(f'Vectors of matrix A4: \n v1: {a4} \n v2: {a44}')
```

The variables $a_i$ and $a_{ii}$ for $i=1:4$ represent the vectors $\lambda_iv_i$ of the matrix $A_i$.


```python
# This function is declared to evaluate the quadratic form associated with a matrix A and a vector v.
def quadratic_form(v, A):
    return A[0,0]*v[0]**2 + (A[0,1] + A[1,0])*v[0]*v[1] + A[1,1]*v[1]**2
```

Information on how to plot 2D math vectors with Matplotlib can be found [here](https://stackoverflow.com/questions/12265234/how-to-plot-2d-math-vectors-with-matplotlib).

```python
# Plotting the previously found vectors on the plane, for each matrix.
# Using the quiver function to create vectors on the plane.
c = 0
for mat in matrices:
    c += 1
    if c == 1:
        # Creating the figure and the vectors.
        plt.figure()
        ax = plt.gca()
        plt.quiver(0, 0, a1[0], a1[1], angles='xy', scale_units='xy', scale=1)
        plt.quiver(0, 0, a11[0], a11[1], angles='xy', scale_units='xy', scale=1)
        # Titles and axes.
        ax.set_title(f'λivi Vectors for A1')
        ax.set_xlim([-5, 5])
        ax.set_ylim([-5, 5])
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        plt.show()
    elif c == 2:
        # Creating the figure and the vectors.
        plt.figure()
        ax = plt.gca()
        plt.quiver(0, 0, a2[0], a2[1], angles='xy', scale_units='xy', scale=1)
        plt.quiver(0, 0, a22[0], a22[1], angles='xy', scale_units='xy', scale=1)
        # Titles and axes.
        ax.set_title(f'λivi Vectors for A2')
        ax.set_xlim([-10, 10])
        ax.set_ylim([-10, 10])
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        plt.show()
    elif c == 3:
        # Creating the figure and the vectors.
        plt.figure()
        ax = plt.gca()
        plt.quiver(0, 0, a3[0], a3[1], angles='xy', scale_units='xy', scale=1)
        plt.quiver(0, 0, a33[0], a33[1], angles='xy', scale_units='xy', scale=1)
        # Titles and axes.
        ax.set_title(f'λivi Vectors for A3')
        ax.set_xlim([-5, 5])
        ax.set_ylim([-5, 5])
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        plt.show()
    elif c == 4:
        # Creating the figure and the vectors.
        plt.figure()
        ax = plt.gca()
        plt.quiver(0, 0, a4[0], a4[1], angles='xy', scale_units='xy', scale=1)
        plt.quiver(0, 0, a44[0], a44[1], angles='xy', scale_units='xy', scale=1)
        # Titles and axes.
        ax.set_title(f'λivi Vectors for A4')
        ax.set_xlim([-5, 5])
        ax.set_ylim([-5, 5])
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        plt.show()
```

### b)

Contour plots with the vectors from the previous section on the same graph.

```python
# Doing exactly the same, but now putting contour plots on the same plot as the vectors.

# Counter to identify each matrix.
c = 0
for mat in matrices:
    
    step = 0.025
    x = np.arange(-10, 10, step)
    y = np.arange(-10, 10, step)

    X, Y = np.meshgrid(x, y)
    # Declaring the quadratic form, which will be different for each matrix.
    Z = mat[0,0]*X**2 + (mat[0,1] + mat[1,0])*X*Y + mat[1,1]*Y**2
    
    c += 1
    
    if c == 1:
        # Creating the contour plot.
        fig, ax = plt.subplots()
        CS = ax.contour(X, Y, Z, [quadratic_form(a1, A1), quadratic_form(a11, A1)])
        ax.clabel(CS, inline=True, fontsize=10)
        ax.set_title(f'A1')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        # Creating the vectors.
        plt.quiver(0, 0, a1[0], a1[1], angles='xy', scale_units='xy', scale=1)
        plt.quiver(0, 0, a11[0], a11[1], angles='xy', scale_units='xy', scale=1)
        plt.show()
    elif c == 2:
        # Creating the contour plot.
        fig, ax = plt.subplots()
        CS = ax.contour(X, Y, Z, [quadratic_form(a2, A2)])
        ax.clabel(CS, inline=True, fontsize=10)
        ax.set_title(f'A2')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        # Creating the vectors.
        plt.quiver(0, 0, a2[0], a2[1], angles='xy', scale_units='xy', scale=1)
        plt.quiver(0, 0, a22[0], a22[1], angles='xy', scale_units='xy', scale=1)
        plt.show()
    elif c == 3:
        # Creating the contour plot.
        fig, ax = plt.subplots()
        CS = ax.contour(X, Y, Z, [quadratic_form(a33, A3), quadratic_form(a3, A3)])
        ax.clabel(CS, inline=True, fontsize=10)
        ax.set_title(f'A3')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        # Creating the vectors.
        plt.quiver(0, 0, a3[0], a3[1], angles='xy', scale_units='xy', scale=1)
        plt.quiver(0, 0, a33[0], a33[1], angles='xy', scale_units='xy', scale=1)
        plt.show()
    elif c == 4:
        # Creating the contour plot.
        fig, ax = plt.subplots()
        CS = ax.contour(X, Y, Z, [quadratic_form(a44, A4), quadratic_form(a4, A4)])
        ax.clabel(CS, inline=True, fontsize=10)
        ax.set_title(f'A4')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        # Creating the vectors.
        plt.quiver(0, 0, a4[0], a4[1], angles='xy', scale_units='xy', scale=1)
        plt.quiver(0, 0, a44[0], a44[1], angles='xy', scale_units='xy', scale=1)
        plt.show()
```

## Part 4

The quadratic form $$q(x,y) =  4x^{2} + 2xy - 3y^{2}$$ is considered.

### a) 

The functions *plot_surface*, *contour*, and *contourf* are used to plot this function.

```python
# Matrix associated with the quadratic form
A = np.array([[4, 1], [1, -3]])
```

*plot_surface*

```python
# Plot using plot_surface

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-1, 1, 0.05)
X, Y = np.meshgrid(x, y)

# Quadratic form in terms of matrix A.
Z = A[0, 0] * X**2 + (A[0, 1] + A[1, 0]) * X * Y + A[1, 1] * Y**2

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title(f'Plot of q(x,y)')
plt.show()
```

*contour*

```python
step = 0.025
x = np.arange(-1.0, 1.0, step)
y = np.arange(-1.0, 1.0, step)

X, Y = np.meshgrid(x, y)
Z = A[0, 0] * X**2 + (A[0, 1] + A[1, 0]) * X * Y + A[1, 1] * Y**2

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=True, fontsize=10)
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.set_title(f'Contour of q(x,y)')
plt.show()
```

*contourf*

```python
inf = -1
sup = 1

delta = 0.025

x = y = np.arange(inf, sup, delta)
X, Y = np.meshgrid(x, y)

# Quadratic form associated with matrix A
Z = A[0, 0] * X**2 + (A[0, 1] + A[1, 0]) * X * Y + A[1, 1] * Y**2

fig1, ax2 = plt.subplots(constrained_layout=True)
CS = ax2.contourf(X, Y, Z, 10, cmap=plt.cm.bone, origin='lower')
ax2.set_title('Contour of q(x,y)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

# Color bar
cbar = fig1.colorbar(CS)
cbar.ax.set_ylabel('Z')
plt.show()
```

### b) 

The contour plot with the eigenvectors on the same plane is obtained.

```python
# Obtain eigenvalues and eigenvectors of the matrix.
l, v = LA.eig(A)

# Create variables λivi for each eigenvalue and eigenvector of the quadratic form.
a5 = l[0] * v[:, 0]
a55 = l[1] * v[:, 1]

step = 0.025
x = np.arange(-10, 10, step)
y = np.arange(-10, 10, step)

X, Y = np.meshgrid(x, y)

# Quadratic form associated with matrix A.
Z = A[0, 0] * X**2 + (A[0, 1] + A[1, 0]) * X * Y + A[1, 1] * Y**2

# Plot the contour of the function.
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z, [quadratic_form(a55, A), quadratic_form(a5, A)])
ax.clabel(CS, inline=True, fontsize=10)
ax.set_title(f'q(x,y)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
# Plot the vectors.
plt.quiver(0, 0, a5[0], a5[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, a55[0], a55[1], angles='xy', scale_units='xy', scale=1)
plt.show()
```
