### Función para hallar tiempo entre intersecciones

```python
def obtenerTiempo(punto_inicial, punto_final, codigo):

    # Inicializa
    np.random.seed(codigo)
    m, n = 13, 7
    p = np.random.rand(m, n)
    sigma = 50

    # Cambia formato
    punto_inicial = np.array([int(i) for i in punto_inicial]).reshape(-1,1)
    punto_final = np.array([int(i) for i in punto_final]).reshape(-1,1)

    if sum(np.diag(cdist(punto_inicial, punto_final, metric='cityblock'))) > 1 or np.all(
            punto_final == punto_inicial) is True:
        return np.inf
    elif punto_inicial[0] > m or punto_inicial[0] < 0 or punto_inicial[1] > n or punto_inicial[1] < 0 :
        return "El punto inicial está por fuera de la malla"
    elif punto_final[0] > m or punto_final[0] < 0 or punto_final[1] > n or punto_final[1] < 0 :
        return "El punto final está por fuera de la malla"

    elif len(codigo) != 9:
        return "Código no válido, no tiene 9 dígitos"
    else:
        c = 4
        for i in range(m):
            for j in range(n):
                if c < len(codigo) - 1:
                    c += 1
                else:
                    c = 5
                if codigo[c-1] != 0:
                    p[i, j] *= codigo[c]
        # Con distribución normal
        Neg = 1
        while Neg:
            mu = p[punto_inicial[0], punto_inicial[1]] + p[punto_final[0], punto_final[1]]
            t = np.random.normal(mu, scale=sigma)
            if t > 0:
                Neg = 0
    return t
```

### Función para dibujar y hallar flujo mínimo

```python
def flujo_mínimo(M: np.array, b: np.array, m_grid = 13, n_grid = 7):

    # Formato y variables iniciales
    b = b.reshape(-1, 1)
    m, n = M.shape
    G = nx.Graph()


    # Crea grafo a partir de Matriz de Distancias
    for i in range(m):
        G.add_node(i + 1)
        for j in range(n):
            valor = M[i, j]
            if valor != np.inf:
                G.add_edge(i + 1, j + 1)
            # else:
            #     print("No se agrega el arco {}->{}".format(i+1,j+1))
    G = G.to_directed()


    # Crea matriz de Adjacencia
    temp_M = np.zeros((m, n))
    indices_no_infinitos = np.where(M != inf)
    temp_M[indices_no_infinitos[0], indices_no_infinitos[1]] = 1
    A = []
    # Agrega por filas
    for i in range(m):
        temp = np.zeros((m, n))
        temp[i, :] = temp_M[i, :]
        temp[:, i] = temp_M[:, i] * -1
        temp[i, i] = 0
        fila = temp.ravel()
        A.append(fila)
    A = np.array(A)

    # Formato para optimización lineal
    M[M == inf] = 0
    c = M.ravel()
    bounds = [(0, None)] * len(c)

    # Solucionar
    res = linprog(c, A_eq=A, b_eq=b, bounds=bounds, method="revised simplex")
    print(res)

    # Crea camino solución
    X = res.x
    costo = res.fun

    X = X.reshape(m_grid*n_grid, m_grid*n_grid )
    result = np.where(X != 0)
    A, B = list(result[0] + 1), list(result[1] + 1)
    A.reverse()
    B.reverse()
    arcos = list(zip(A, B))

    # for arco in arcos:
    #     if not g.has_edge(arco[0],arco[1]):
    #         print("el arco {}->{} no deberia estar en la solución pues no está en el grafo".format(arco[0],arco[1]))

    # Etiquetas de nodos
    pos = create_grid_layout(m_grid, n_grid)
    labels_pos = {k: np.array([0.3, 0.25]) + np.array(v) for k, v in pos.items()}

    # Restaurantes
    x, y = [2,1], [1,5]
    plt.scatter(x, y, color = 'darkred', marker = 's', s = 120, zorder = 90)

    # Casa
    x, y = [5], [4]
    plt.scatter(x, y, color = 'steelblue', marker = 'p', s = 170, zorder = 90)

    # Universidad
    x, y = [11], [1]
    plt.scatter(x, y, color = 'gold', marker = 'd', s = 160, zorder = 90)
 
    # Dibuja grafo
    labels = nx.draw_networkx_labels(G, pos=labels_pos, font_color="grey", font_size=10)
    nx.draw_networkx(G, pos, node_color='lightgray', node_size=12, with_labels=False)
    nx.draw_networkx_edges(G, pos, edgelist=arcos, edge_color='limegreen', width=4, arrowsize = 6)

    plt.show()

    return costo, X, G

# Código estandar
def create_grid_layout(n, m):
    dictionary = {}
    for node in range(n * m):
        index = np.unravel_index(node, (n, m))
        dictionary[node + 1] = index
    return dictionary
```

### a) Planteamiento matemático del problema a resolver

En el problema a resolver, se tienen 91 nodos, en donde cada nodo representa una intersección entre diferentes calles. Las calles representan los arcos del problema. Como cada calle tiene asociado un costo $c_{ij}$, que representa el tiempo que se tarda en recorrer esta calle particular, se plantea encontrar las calles que se deben recorrer para minimizar el viaje entre la casa y la universidad. Estas calles que indican cuál es la ruta óptima para el viaje están representadas por los flujos $x_{ij}$. 

Entonces, la función objetivo de este problema de optimización es la suma del producto de los costos de cada arco $c_{ij}$ (tiempo de viaje entre cada intersección) y los flujos $x_{ij}$. El valor de estos flujos va ser 0 o 1 dependiendo de si se está tomando en cuenta ese arco (calle) como parte de la ruta óptima.

Las restricciones que se tienen para este problema de optimización tienen que ver con el flujo neto que hay entre cada nodo, o en este caso, cada intersección. Este flujo neto para el primer nodo (punto de partida) tiene que ser 1 pues de este nodo solo hay un flujo saliendo, tiene que ser -1 para el último nodo (punto de llegada) pues a este solo está entrando un flujo y tiene que ser 0 para el resto de nodos intermedios. En otras palabras, la cantidad de flujo que entra a un nodo tiene que ser igual a la cantidad de flujo que sale del mismo. Esta cantidad de flujo neto está representada por las variables $b_i$.

En resumen, se quiere minimizar el tiempo que toma viajar de un punto de partida particular (casa) a un punto de llegada (universidad), sabiendo que el flujo neto de cada intersección debe ser 0 y que solo hay un flujo del punto de partida y un flujo negativo en el punto de llegada. La formulación matemática del problema se muestra a continuación:

min $f(x) = \sum_{i=1}^{91}\sum_{j=1}^{91}c_{ij}x_{ij}$\
s.a.\
$\sum_{j=1}^{91}x_{ij} - \sum_{k=1}^{91}x_{ki} = b_i$

donde 

$i = 1, ... ,91 $ 

y 

$b_{partida} = 1$ (en este caso $b_{partida} = b_{1}$) \
$b_{llegada} = -1$ (en este caso $b_{llegada} = b_{91}$) \
$b_i = 0$ para los demás nodos.

### b) Construcción de la matriz de adyacencia del problema de optimización

Información sobre cómo voltear los valores dentro de una matriz se encontró [aquí](https://numpy.org/doc/stable/reference/generated/numpy.flip.html).

```python
# Se crea la matriz de adyacencia y una matriz auxiliar que ayudará a señalar el número del nodo correspondiente a cada
# intersección dentro del mapa.
A = np.zeros((91,91))
aux = np.zeros((13,7))

# Se llena la matriz auxiliar con números ascendentes, para que cada componente de esta matriz represente el número del
# nodo correspondiente a esa intersección.
c = 1
for i in range(0,np.shape(aux)[0]):
    for j in range(np.shape(aux)[1]):
        aux[i,j] = c
        c+=1
print(f'Matriz auxiliar : \n {aux}')

# Se le da vuelta a la transpuesta de la matriz auxiliar original para que los componentes cuadren con el sistema de 
# coordenadas descrito en la guía. 
aux2 = np.flip(aux.transpose(),0)
print(f'\nMatriz auxiliar volteada y transpuesta (aux2):  \n {aux2}')

# Se recorre la matriz auxiliar aux2 para llenar la matriz de adyacencia. Se empiezan los índices en 1 para que cuadre con
# la numeración asignada a los nodos. Como consecuencia de esto, se le debe restar 1 a cada una de las posiciones indexadas 
# dentro de las matrices A y aux2.
for i in range(1, np.shape(aux2)[0]+1):
    for j in range(1, np.shape(aux2)[1]+1):
        # Si la componente actual tiene vecino abajo
        if i+1<=np.shape(aux2)[0]:
            # Se determina a qué índice de la matriz de adyacencia corresponde la componente actual.
            x = int(aux2[i-1,j-1])
            y = int(aux2[i-1+1,j-1])
            # En la matriz de adyacencia se coloca un 1 en la posición del vecino de abajo.
            A[x-1,y-1] = 1
        # Si la componente actual tiene vecino arriba
        if i-1>=1:
            # Se determina a qué índice de la matriz de adyacencia corresponde la componente actual.
            x = int(aux2[i-1,j-1])
            y = int(aux2[i-1-1,j-1])
            # En la matriz de adyacencia se coloca un 1 en la posición del vecino de arriba.
            A[x-1,y-1] = 1
        # Si la componente actual tiene vecino a la derecha
        if j+1<=np.shape(aux2)[1]:
            # Se determina a qué índice de la matriz de adyacencia corresponde la componente actual.
            x = int(aux2[i-1,j-1])
            y = int(aux2[i-1,j-1+1])
            # En la matriz de adyacencia se coloca un 1 en la posición del vecino de la derecha.
            A[x-1,y-1] = 1
        # Si la componente actual tiene vecino a la izquierda
        if j-1>=1:
            # Se determina a qué índice de la matriz de adyacencia corresponde la componente actual.
            x = int(aux2[i-1,j-1])
            y = int(aux2[i-1,j-1-1])
            # En la matriz de adyacencia se coloca un 1 en la posición del vecino de la izquierda.
            A[x-1,y-1] = 1

print(f'\nMatriz de adyacencia A: \n {A}')
```

### c) Construcción de la matriz M, en donde cada valor distinto de 0 en la matriz de adyacencia es reemplazado por el costo/tiempo del arco correspondiente. Si el arco no existe, el valor correspondiente a este se reemplaza por $\infty$.


Información sobre cómo acceder a los índices de un valor específico dentro de una matriz se encontró [aquí](https://thispointer.com/find-the-index-of-a-value-in-numpy-array/).

Información sobre cómo construir una matriz de tuplas se encontró [aquí](https://stackoverflow.com/questions/8024750/numpy-cant-build-a-2-d-array-of-tuples).

```python
# Se crea la matriz M.
M = A.copy()

# Se crea una matriz de tuplas, en donde cada posición representará un punto en el sistema de coordenadas descrito 
# en la guía. 
C = np.empty((7,13),object)
for i in range(np.shape(C)[0]):
    for j in range(np.shape(C)[1]):
        C[i,j] = (j,i)

# Se le da vuelta a esta matriz para que las posiciones de las coordenadas cuadren con el sistema descrito en la guía. 
tuplas = np.flip(C,0)

# Se recorre la matriz M para ir reemplazando los valores necesarios. Nuevamente se empiezan los índices en 1 y como
# consecuencia, se le resta 1 a cada una de las posiciones indexadas dentro de la matriz M.
for i in range(1,len(M)+1):
    for j in range(1,len(M)+1):
        # Se reemplazarán los 1s de la matriz de adyacencia por los costos asociados al par de nodos correspondiente.
        if M[i-1,j-1] == 1:
            # La matriz 'fuente' es una matriz llena de infinitos a excepción de las posiciones en donde la matriz 'aux2'
            # tiene el valor igual al índice del nodo fuente de donde se encontró el 1 en la matriz de adyacencia. Lo mismo
            # sucede con la matriz 'destino'.
            fuente = np.where((aux2 == i), i, inf)
            destino = np.where((aux2 == j), j, inf)
            # Dentro de las matrices 'fuente' y 'destino' se determina en qué indices hay un valor diferente de infinito 
            # (siempre va a haber solo 1 valor diferente de infinito). Estos índices representan la posición dentro de la
            # matriz de tuplas en las que se va a encontrar la coordenada correspondiente tanto al nodo fuente como el 
            # nodo destino. 
            f = np.where(fuente!=inf)
            d = np.where(destino!=inf)
            # Se extrae la coordenada correspondiente a los índices encontrados previamente de la matriz de tuplas tanto
            # para nodo origen como nodo destino.
            origen = tuplas[f[0],f[1]][0]
            destino = tuplas[d[0],d[1]][0]
            # Se obtiene el tiempo entre el par de coordenadas de cada nodo y se reemplaza el 1 que estaba en la matriz
            # de adyacencia por el tiempo correspondiente entre los dos nodos. 
            M[i-1,j-1] = np.round(obtenerTiempo(origen,destino,[2,0,1,7,3,0,4,1,8]),2)
        # No hay conexiones entre un mismo nodo, entonces su valor en la matriz M se pone como un tiempo infinito.
        if j-1==i-1:
            M[i-1,j-1] = np.inf
        # El resto de valores que no son ni flujos ni infinitos se convierten en infinito.
        elif M[i-1,j-1] == 0:
            M[i-1,j-1] = np.inf
                
print(f'Matriz M: \n {M}')
```

### d) Construcción del vector $s$ que representa los flujos netos dentro de cada nodo.

```python
# Creación del vector columna.
n = np.shape(A)[0]
s = np.zeros((n,1))

# Se empieza el índice en 1 para seguir con la notación usada previamente. 
for k in range(1,len(s)+1):
    # Como se va a encontrar la ruta mínima entre la casa y la universidad, se pone un 1 en el nodo de la casa (nodo 40), 
    # pues es el nodo de partida, y se pone un -1 en el nodo de la universidad, que es el nodo de llegada y corresponde al
    # nodo 79. El resto de componentes es cero por la forma en que se inicializó el vector s.
    if k == 40:
        s[k-1] = 1
    elif k==79:
        s[k-1] = -1
```

### e) Se usa la función *flujo_mínimo* para encontrar la ruta más corta entre la casa y la universidad. Se grafica esta ruta más corta.

```python
costo, X, G = flujo_mínimo(M,s)
```

De acuerdo al resultado obtenido, las flechas verdes indican la ruta más corta entre la casa y la universidad y se tiene que el tiempo que toma recorrer esta ruta óptima es de 560.47 segundos, o **9.34 minutos**.

```python
# Creación del vector columna.
n3 = np.shape(A)[0]
s3 = np.zeros((n3,1))

# Se empieza el índice en 1 para seguir con la notación usada previamente. 
for k in range(1,len(s3)+1):
    # Como se va a encontrar la ruta mínima entre la universidad y la casa, se pone un 1 en el nodo de la universidad
    # (nodo 79), pues es el nodo de partida, y se pone un -1 en el nodo de la casa, que es el nodo de llegada y corresponde al
    # nodo 40. El resto de componentes es cero por la forma en que se inicializó el vector s3.
    if k == 79:
        s3[k-1] = 1
    elif k==40:
        s3[k-1] = -1
```
