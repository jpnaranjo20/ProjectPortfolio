# Central Limit Theorem


Let $X_i$ with $i=1, \dots, n$ be i.i.d. random variables, with uniform distribution between 2 and 4 ($U[2,4]$).

Let $Y_n = \frac{S_n - n\mu}{\sigma \sqrt{n}}$, where $S_n = \sum_{i=1}^{n} X_i$, $\mu = E[X_i]$, and $\sigma^2 = \text{var}(X_i)$. Note that $\mu$ and $\sigma^2$ can be calculated theoretically.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### a) Generate 5000 realizations of the variable $X_1$. Calculate $Y_n$ and plot the histogram of $Y_n$ (there must be 5000 realizations of this random variable). Assume $n = 1$. 

```python
# Function to convert an array x into a column vector.
def to_column_vector(x):
    return np.asarray(np.matrix(x)).T

# Function to calculate the mean of a uniform distribution between a and b.
def mu_uniform(a, b):
    return (1/2)*(b+a)

# Function to calculate the variance of a uniform distribution between a and b.
def var_uniform(a, b):
    return (1/12)*(b-a)**2

# Function to calculate Yn
def calculate_Yn(n):

    # Number of samples for each RV.
    samples = 5000

    # Define limits of the uniform distribution
    a = 2
    b = 4

    # Create a matrix where each column will have a vector with realizations of each random variable.
    random_variables = np.zeros((samples,n))

    # Fill this matrix depending on the number of random variables to be used.
    # Each column is a different RV with uniform distribution between a and b.
    for i in range(n):
        random_variables[:,i] = to_column_vector(np.random.uniform(a, b, samples)).T
        
    # Create the vector that sums all random variables in the matrix.
    Sn = to_column_vector(np.sum(random_variables, axis = 1))

    # Calculate the vector Yn according to the formula. This turns out to be a vector of the same size as the X_i variables.
    Yn = (Sn - n*mu_uniform(a, b))/(np.sqrt(var_uniform(a, b)*n))

    return Yn
```

```python
# Plotting the histogram of Yn for n = 1
n = 1
plt.figure()
m, bins, patches = plt.hist(calculate_Yn(n), 20)
plt.grid()
plt.title(r'Histogram of $Y_n$ ' + f'with n={n}', fontsize=18)
plt.xlabel(r'$Y_n$', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.show()
```

### b) Assume $n = 2$

For each of the random variables $X_1$ and $X_2$, generate 5000 realizations. Calculate $Y_n$, and plot the histogram of $Y_n$ (there must be 5000 realizations of this random variable).

```python
# Plotting the histogram of Yn for n = 2
n = 2
plt.figure()
m, bins, patches = plt.hist(calculate_Yn(n), 20)
plt.grid()
plt.title(r'Histograma de $Y_n$ ' + f'con n={n}', fontsize=18)
plt.xlabel(r'$Y_n$', fontsize=16)
plt.ylabel('Frecuencia', fontsize=16)
plt.show()
```

### c) Now repeat the previous procedure for $n = 3, 5, 10, 100, 1000, 100000$.

Analyze the results and explain them in terms of the central limit theorem.

```python
# Plotting the histogram of Yn for n = 2
n_values = [3, 5, 10, 100, 1000, 100000]
for n in n_values:
    plt.figure()
    m, bins, patches = plt.hist(calculate_Yn(n), 20)
    plt.grid()
    plt.title(r'Histogram of $Y_n$ ' + f'with n={n}', fontsize=18)
    plt.xlabel(r'$Y_n$', fontsize=16)
    plt.ylabel('Frequency', fontsize=16)
    plt.show()
```

As it can be noticed, as the value of $n$ increased, the histogram of $Y_n$ started to take a shape closer and closer to the histogram of a random variable following a standard normal distribution (mean equal to zero and variance equal to 1). It is evident that the histogram of the random variable $Y_n$ quickly began to converge to the standard normal histogram, as from $n=2$ onwards it already started to take a shape similar to the characteristic bell curve of the normal distribution. It can also be observed that approximately around $n=100$, the shape of the histogram of $Y_n$ already began to be practically identical to that of the standard normal histogram, and for $n$ s greater than this, it was even more similar.

It can be concluded that the experiment carried out in this section provided a clear demonstration of the central limit theorem, as by adding more independent identically distributed random variables to generate a specific $S_n$, the random variable $Y_n = \frac{S_n - n\mu}{\sigma \sqrt{n}}$ approached more and more to follow a standard normal distribution, which is exactly what the theorem indicates. In this case, a relatively rapid convergence to the standard normal distribution was observed, and this could be due to the fact that the shape of the uniform distribution characterizing our random variables $X_i$ is not substantially different from the bell curve shape, compared to other distributions such as the exponential distribution, for example. It is possible that for a collection of iid (independent, identically distributed) random variables following a distribution whose shape is very different from that of a bell curve, the distributional convergence of the random variable $Y_n$ to a standard normal distribution is "slower" in this case, in the sense that a larger number of random variables $X_i$ must be added before $Y_n$ starts to converge in distribution to a standard normal distribution.
