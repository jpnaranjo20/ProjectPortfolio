```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import sqrtm
```

```python
Cy = np.array([[25, 1], [1, 9]])
Cx = np.array([[1, 0.1], [0.1, 2]])
```

```python
# Vectores propios de Cx
valsCx, vectsCx = np.linalg.eig(Cx) # Ya salen normalizados

# Se crea matriz Lambda con los valores propios
LambdaCx = np.array([[valsCx[0], 0], [0, valsCx[1]]])

# Se crea la matriz G, con G = Cx^1/2
G = np.matmul(np.matmul(vectsCx, np.sqrt(LambdaCx)), vectsCx.T)
print(f'Matriz G calculada con la fórmula: \n {G}')
G2 = sqrtm(Cx)
print(f'\nMatriz G calculada con función sqrtm: \n {G2}')
```

```python
# Vectores propios de Cy
valsCy, vectsCy = np.linalg.eig(Cy) # Ya salen normalizados

# Se crea matriz Lambda con los valores propios
LambdaCy = np.array([[valsCy[0], 0], [0, valsCy[1]]])
# Se crea matriz Lambda con 1/valores propios
LambdaCy = np.array([[1/valsCy[0], 0], [0, 1/valsCy[1]]])

# Se crea la matriz H, con H = Cx^-1/2
H = np.matmul(np.matmul(vectsCy, np.sqrt(LambdaCy)), vectsCy.T)
print(f'Matriz H calculada con la fórmula = \n {H}')
H2 = sqrtm(Cy)
print(f'\nMatriz H calculada con función sqrtm = \n {H2}')
```

```python
A = np.matmul(G, H)
print(f'Matriz A = \n {A}')

print(f'\nSe confirma que la matrix Cx está dada por ACyA^T= \n {np.matmul(np.matmul(A, Cy), A.T)}')
```
