```python
import numpy as np
```

```python
A = np.array([[25, 20], [20, 9]])
B = np.array([[3, -2], [-2, 3]])
D = np.array([[4, 0, 0], [0, 3, 0], [0, 0, 5]])

# Obtenemos los valores propios de cada matriz para determinar si pueden ser matrices de covarianza
valsA, vectsA = np.linalg.eig(A)
valsB, vectsB = np.linalg.eig(B)
valsD, vectsD = np.linalg.eig(D)

print(f'Valores propios de A: {valsA}')
print(f'Valores propios de B: {valsB}')
print(f'Valores propios de D: {valsD}')
```
