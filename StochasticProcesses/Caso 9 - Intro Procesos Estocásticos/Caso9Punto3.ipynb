```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

# Punto 3 b)

```python
# Vector de tiempo
t = np.arange(1,51)

# Definición del proceso estocástico
def proc_estoc(t, A, theta):
    f = 0.1 # Hz
    return A*np.cos(2*np.pi*f*t + theta)

# Creamos las variables aleatorias uniformes A y theta.
A = np.random.uniform(0, 5, 1000)
theta = np.random.uniform(0, np.pi, 1000)

# Como vamos a crear 10 realizaciones del proceso estocástico, vamos a guardarlas en una matriz de 50x10.
realizaciones = np.zeros((len(t), 10))
n, m = np.shape(realizaciones)

# Creamos cada realización del proceso estocástico. Cada columna de la matriz representa una realización.
# En cada componente de cada columna de la matriz creamos una realización del proceso estocástico en ese tiempo con una cantidad aleatoria de amplitud y desfase.
for j in range(m):
    for i in range(n):
        amp = random.choice(A)
        desfase = random.choice(theta)
        realizaciones[i, j] = proc_estoc(i, amp, desfase)

# Gráficas
plt.figure(figsize=(12,8))
plt.grid()
plt.title(f'Realizaciones del proceso estocástico $X(t)$', fontsize=16)
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$X(t)$', fontsize=16)

for i in range(m):
    plt.plot(t, realizaciones[:,i], label=f'Realización {i+1}')

plt.legend()
plt.savefig('realizaciones.eps', format='eps')
plt.show()
```
