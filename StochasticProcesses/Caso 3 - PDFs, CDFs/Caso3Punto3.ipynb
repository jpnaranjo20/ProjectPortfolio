# Punto 3
Se tiene la siguiente función de densidad de probabilidad con parámetros $-1 < \rho < 1$, y $\mu_x$, $\mu_y \in \mathbb{R}$.

$$f_{X,Y}(x,y) = \frac{1}{2\pi\sqrt{1-\rho^2}}\exp{\frac{-((x-\mu_x)^2-2\rho(x-\mu_x)(y-\mu_y) + (y-\mu_y)^2)}{2(1-\rho^2)}}$$

para todo $x,y \in \mathbb{R}$.

## a)
Elija unos valores específicos para $\mu_x$ y $\mu_y$. Grafique las superficies de contorno de $f_{X,Y}(x,y)$ para varios valores de $\rho$ (recuerde que $-1 < \rho < 1$).

````python
mux = 0
muy = 0
rho = 0.5
rhos = [-0.9, -0.7, -0.5, -0.3, -0.1, 0, 0.1, 0.3, 0.5, 0.7, 0.9]

step = 0.025
x = np.arange(-3, 3, step)
y = np.arange(-3, 3, step)

X, Y = np.meshgrid(x, y)

for rho in rhos:
    Z = (1/(2*np.pi*np.sqrt(1-rho**2)))*np.exp(-((X-mux)**2-2*rho*(X-mux)*(Y-muy)+(Y-muy)**2)/(2*(1-rho**2)))

    fig, ax = plt.subplots(figsize=(8,8))
    CS = ax.contour(X, Y, Z)
    ax.clabel(CS, inline=True, fontsize=10)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title(r'Curvas de contorno de $f_{X,Y}(x,y)$ con $\rho$ = ' + f'{rho}')
    ax.grid()
    plt.figure() 
````

Para esto, grafique las muestras en ```datosCaso3punto3.txt``` en el plano $x$-$y$, y traslape la gráfica del contorno de la función de densidad candidata.

````python
# Se importan los datos del .txt
filename = 'datosCaso3punto3.txt'
col1_data = np.loadtxt(filename)[:, 0]
col2_data = np.loadtxt(filename)[:, 1]

mu_x_candidato = -1
mu_y_candidato = 5
rho_candidato = -0.65

step = 0.025
x = np.arange(-4.5, 3, step)
y = np.arange(1.5, 8.5, step)

X1, Y1 = np.meshgrid(x, y)
Z1 = (1/(2*np.pi*np.sqrt(1-rho_candidato**2)))*np.exp(-((X1-mu_x_candidato)**2-2*rho_candidato*(X1-mu_x_candidato)*(Y1-mu_y_candidato)+(Y1-mu_y_candidato)**2)/(2*(1-rho_candidato**2)))

fig, ax = plt.subplots(figsize=(8,8))
CS = ax.contour(X1, Y1, Z1, colors=['black'])
ax.clabel(CS, inline=True, fontsize=10)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title(r'Muestras de los datos traslapados con las curvas de nivel de una función '+ "\n" + r'$f_{X,Y}(x,y)$ con parámetros $\rho$ = ' + f'{rho_candidato} ' + r'$\mu_x$ = ' + f'{mu_x_candidato} ' + r'y $\mu_y$ = ' + f'{mu_y_candidato}')
ax.grid()
plt.scatter(col1_data, col2_data)
plt.figure()
````
