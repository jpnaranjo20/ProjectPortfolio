```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Punto 1

```python
valores_x = [-3, -1, 0, 1, 2]

def px(x):
    if x==-3 or x == -1 or x==0 or x==1 or x==2:
        resp = (1/1.7542)*np.exp(-x**2)
    else:
        resp = 0
    
    return resp
```

## a)

```python
def valorEsperado(valores):
    resp = 0
    for val in valores:
        resp += val*px(val)

    return resp

EX = valorEsperado(valores_x)

print(f'El valor esperado de X es {EX}')
```

```python
def varianza(valorEsperado, valores):
    EXcuad = 0
    for val in valores:
        EXcuad += (val**2)*px(val)
    
    resp = EXcuad - (EX)**2

    return resp

print(f'La varianza de X es: {varianza(EX, valores_x)}')
```

## b) Encuentre la PMF de la variable aleatoria $Z = e^{-(X-E[X])^2}$

```python
def Z(X):
    return np.exp(-(X-EX)**2)
```

```python
valores_z = []
pz = []

for val in valores_x:
    valores_z.append(Z(val))
    pz.append(px(val))

d = {'X': valores_x, 'Z': valores_z, 'p_z(Z)': pz}
df = pd.DataFrame(data=d)

print(f'La funci칩n de masa de probabilidad de la variable aleatoria Z est치 descrita en la siguiente tabla: \n')
print(f'{df} \n')
print(f'La suma de la PMF de Z es igual a {np.sum(pz)}')
```
