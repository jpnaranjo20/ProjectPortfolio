# Punto 5

Sean $X_i$ con $i=1, \dots, n$ variables aleatorias i.i.d, con distribución uniforme entre 2 y 4 ($U[2,4]$).

Sea $Y_n = \frac{S_n - n\mu}{\sigma \sqrt{n}}$, con $S_n = \sum_{i=1}^{n} X_i$, $\mu = E[X_i]$ y $\sigma^2 = \text{var}(X_i)$. Tenga en cuenta que $\mu$ y $\sigma^2$ se pueden calcular teóricamente.

## a) Asuma $n = 1$.
Genere 5000 realizaciones de la variable $X_1$. Calcule $Y_n$ y grafique el histograma de $Y_n$ (tienen que haber 5000 realizaciones de esta variable aleatoria).

```python
# Función para convertir un arreglo x en un vector columna.
def vector(x):
    return np.asarray(np.matrix(x)).T

# Función para hallar la media de una distribución uniforme entre a y b.
def mu_unif(a, b):
    return (1/2)*(b+a)

# Función para hallar la varianza de una distribución uniforme entre a y b.
def var_unif(a, b):
    return (1/12)*(b-a)**2

# Función para calcular Yn
def calcYn(n):

    # Número de muestras de cada VA.
    samples = 5000

    # Definir límites de la distribución uniforme
    a = 2
    b = 4

    # Se crea una matriz en la que cada columna habrá un vector con las realizaciones de cada variable aleatoria.
    VAs = np.zeros((samples,n))

    # Se llena dicha matriz dependiendo del número de VAs que se vayan a usar.
    # Cada columna es una VA diferente de distribución uniforme entre a y b.
    for i in range(n):
        VAs[:,i] = vector(np.random.uniform(a, b, samples)).T
        
    # Se crea el vector que suma todas las variables aleatorias de la matriz.
    Sn = vector(np.sum(VAs, axis = 1))

    # Se calcula el vector Yn de acuerdo a la fórmula. Este resulta ser un vecor del mismo tamaño que las variables X_i
    Yn = (Sn - n*mu_unif(a, b))/(np.sqrt(var_unif(a, b)*n))

    return Yn
```

```python
# Se grafica el histograma de Yn para n = 1
n = 1
plt.figure()
m, bins, patches = plt.hist(calcYn(n), 20)
plt.grid()
plt.title(r'Histograma de $Y_n$ ' + f'con n={n}', fontsize=18)
plt.xlabel(r'$Y_n$', fontsize=16)
plt.ylabel('Frecuencia', fontsize=16)
```

## b) Asuma $n = 2$
Para cada una de las variables aleatorias $X_1$ y $X_2$, genere 5000 realizaciones. Calcule $Y_n$, y grafique el histograma de $Y_n$ (tienen que haber 5000 realizaciones de esta variable aleatoria).

```python
# Se grafica el histograma de Yn para n = 2
n = 2
plt.figure()
m, bins, patches = plt.hist(calcYn(n), 20)
plt.grid()
plt.title(r'Histograma de $Y_n$ ' + f'con n={n}', fontsize=18)
plt.xlabel(r'$Y_n$', fontsize=16)
plt.ylabel('Frecuencia', fontsize=16)
```

## c) Ahora repita el procedimiento anterior para $n = 3, 5, 10, 100, 1000, 100000$.
Analice los resultados y explíquelos en términos del teorema del límite central

```python
# Se grafica el histograma de Yn para n = 2
enes = [3, 5, 10, 100, 1000, 100000]
for n in enes:
    plt.figure()
    m, bins, patches = plt.hist(calcYn(n), 20)
    plt.grid()
    plt.title(r'Histograma de $Y_n$ ' + f'con n={n}', fontsize=18)
    plt.xlabel(r'$Y_n$', fontsize=16)
    plt.ylabel('Frecuencia', fontsize=16)
```
