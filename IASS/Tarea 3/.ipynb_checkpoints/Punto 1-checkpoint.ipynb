```python
from scipy.io import wavfile
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, fftshift, ifft
from numpy import linalg as LA
from scipy.io.wavfile import write
import winsound
import wavio
from playsound import playsound
```

# Punto 1
## a)
Gráfica de la señal de voz en el dominio del tiempo con su respectiva transformada de Fourier.

```python
samprate1, s = wavfile.read('borgesRuido.wav')
s = s/32767 # El formato de s es de un número entero de 16 bits, que tiene un rango entre -32768 a 32767.

length1 = s.shape[0]/samprate1 # Duración en segundos
duration = int(length1) # Duración en segundos enteros

t = np.linspace(0, length1, s.shape[0]) 

# Returns evenly spaced numbers over a specified interval.
# Returns s.shape[0] evenly spaced samples, calculated over the interval [0, length1].

plt.plot(t, s)
plt.grid()
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title("Señal $s(t)$ en el dominio del tiempo")

N = samprate1*duration
fouriers = fft(s)
f1 = fftfreq(s.shape[0], 1/samprate1) # Window length: s.shape[0], Sample spacing: 1/samprate1 

plt.figure()

plt.plot(f1, np.abs(fouriers), color='g')
plt.grid()
plt.xlabel('$f$ (Hz)')
plt.ylabel('|$S(f)$|')
plt.title("Transformada de Fourier de $s(t)$")

playsound('borgesRuido.wav')
```

## b)
Diseño de un filtro tipo FIR de orden $N$ por el método de las ventanas visto en clase. Se grafican los coeficientes del filtro contra $n$ con $n=0:N$, y su respuesta en frecuencia en Hz

```python
# Paso 1 - encontrar Fc (Hz/muestra)

fc1 = 2900 # Hz
fc2 = 3100 # Hz
fcpi = 0.5*samprate1 # Hz

Fc1 = (2*fc1/samprate1)*np.pi # Hz/muestra
print(f'\nFc1 = {Fc1} Hz/muestra')

Fc2 = (2*fc2/samprate1)*np.pi # Hz/muestra
print(f'\nFc2 = {Fc2} Hz/muestra')

Fcpi = (2*fcpi/samprate1)*np.pi # Hz/muestra
print(f'\nFcpi = {Fcpi} Hz/muestra')
```

```python
# Paso 2 - encontrar los b_k

# Se va a construir un rechaza banda que quite las frecuencias de 3000 Hz. Esto se implementa con la suma/resta de tres
# filtros pasabajas.

N = 30 # Orden del filtro

def crear_filtro(N):
    
    def h1(k):
        if k == N/2:
            return (Fc1/np.pi)
        elif k <= N and k >= 0:
            return (np.sin(Fc1*(k-N/2)))/(np.pi*(k-N/2))
        else:
            return 0

    def h2(k):
        if k == N/2:
            return (Fc2/np.pi)
        elif k <= N and k >= 0:
            return (np.sin(Fc2*(k-N/2)))/(np.pi*(k-N/2))
        else:
            return 0

    def hpi(k):
        if k == N/2:
            return (Fcpi/np.pi)
        elif k <= N and k >= 0:
            return (np.sin(Fcpi*(k-N/2)))/(np.pi*(k-N/2))
        else:
            return 0
    
    hk1 = []
    hk2 = []
    hkpi = []

    for i in range(0, N):
        hk1.append(h1(i))
        hk2.append(h2(i))
        hkpi.append(hpi(i))

    # hk = hk1 + hkpi - hk2

    hparcial = np.add(hk1, hkpi)

    hk = np.subtract(hparcial, hk2)
    
    return hk

filtro = crear_filtro(N)

plt.plot(range(0,N), filtro)
plt.grid()
plt.xlabel('$n$')
plt.ylabel('$h_k$')
plt.title("Coeficientes $h_k$ del filtro FIR con " + f"$N={N}$")

fourierhk = fft(filtro)
fhk = fftfreq(len(filtro), 1/samprate1)

plt.figure()

plt.plot(fhk, np.abs(fourierhk), color='g')
plt.grid()
plt.xlim()
plt.xlabel('$f$ (Hz)')
plt.ylabel('|$H_k(f)$|')
plt.title("Transformada de Fourier de los coeficientes $h_k$ del filtro FIR con " + f"$N={N}$")
```

## c) 
Gráfica de la señal filtrada, con su respectiva respuesta en frecuencia

```python
# Paso 3 - Implementar la ecuación de diferencias

def filtrar_señal(N, filtro):
    
    def w(n):
        if n >= 0 and n <= N:
            return 1
        else:
            return 0

    def b(k):
        return filtro[k]*w(k)

    def y(n, x):
        y = 0
        for k in range(0, N):
            y = y + b(k)*x[n-k]

        return y

    filtrada = []

    for n in range(len(s)):
        f = y(n,s)
        filtrada.append(f)
        
    return filtrada
```
