## b)
Dise침o de un filtro tipo FIR de orden $N$ por el m칠todo de las ventanas visto en clase. Se grafican los coeficientes del filtro contra $n$ con $n=0:N$, y su respuesta en frecuencia en Hz

```python
# Paso 1 - encontrar Fc (Hz/muestra)

fc1 = 2900 # Hz
fc2 = 3100 # Hz
fcpi = 0.5*samprate1 # Hz

Fc1 = (2*fc1/samprate1)*np.pi # Hz/muestra
print(f'\nFc1 = {Fc1} Hz/muestra')

Fc2 = (2*fc2/samprate1)*np.pi # Hz/muestra
print(f'\nFc2 = {Fc2} Hz/muestra')

Fcpi = (2*fcpi/samprate1)*np.pi # Hz/muestra
print(f'\nFcpi = {Fcpi} Hz/muestra')
```

```python
# Paso 2 - encontrar los b_k

# Se va a construir un rechaza banda que quite las frecuencias de 3000 Hz. Esto se implementa con la suma/resta de tres
# filtros pasabajas.

N = 30 # Orden del filtro

def crear_filtro(N):
    
    def h1(k):
        if k == N/2:
            return (Fc1/np.pi)
        elif k <= N and k >= 0:
            return (np.sin(Fc1*(k-N/2)))/(np.pi*(k-N/2))
        else:
            return 0

    def h2(k):
        if k == N/2:
            return (Fc2/np.pi)
        elif k <= N and k >= 0:
            return (np.sin(Fc2*(k-N/2)))/(np.pi*(k-N/2))
        else:
            return 0

    def hpi(k):
        if k == N/2:
            return (Fcpi/np.pi)
        elif k <= N and k >= 0:
            return (np.sin(Fcpi*(k-N/2)))/(np.pi*(k-N/2))
        else:
            return 0
    
    hk1 = []
    hk2 = []
    hkpi = []

    for i in range(0, N):
        hk1.append(h1(i))
        hk2.append(h2(i))
        hkpi.append(hpi(i))

    # hk = hk1 + hkpi - hk2

    hparcial = np.add(hk1, hkpi)

    hk = np.subtract(hparcial, hk2)
    
    return hk

filtro = crear_filtro(N)

plt.plot(range(0,N), filtro)
plt.grid()
plt.xlabel('$n$')
plt.ylabel('$h_k$')
plt.title("Coeficientes $h_k$ del filtro FIR con " + f"$N={N}$")

fourierhk = fft(filtro)
fhk = fftfreq(len(filtro), 1/samprate1)

plt.figure()

plt.plot(fhk, np.abs(fourierhk), color='g')
plt.grid()
plt.xlim()
plt.xlabel('$f$ (Hz)')
plt.ylabel('|$H_k(f)$|')
plt.title("Transformada de Fourier de los coeficientes $h_k$ del filtro FIR con " + f"$N={N}$")
```
